```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Initialize empty lists to store the scraped data
name = []
price = []

# Iterate through pages 1 to 25
for j in range(1, 26):
    # Update the URL to include the page number
    URL = f"https://www.amazon.in/s?k=whey+protein&i=todays-deals&crid=K8YF2KWBRM5A&sprefix=%2Ctodays-deals%2C232&ref=nb_sb_ss_recent_2_0_recent&page={j}"
    
    HEADERS = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36',
        'Accept-Language': 'en-US, en;q=0.5'
    }
    
    # Get the webpage content
    webpage = requests.get(URL, headers=HEADERS)
    soup = BeautifulSoup(webpage.content, "html.parser")
    
    # Find the product cards
    company = soup.find_all('div', class_="puis-card-container s-card-container s-overflow-hidden aok-relative puis-include-content-margin puis puis-v2fl5pkubaqu126k6zseo6li6q s-latency-cf-section puis-card-border")
    
    # Scrape product names and prices from the current page
    for i in company:
        # Scrape product name
        product_name = i.find('h2', class_="a-size-mini a-spacing-none a-color-base s-line-clamp-2")
        if product_name:
            name.append(product_name.text.strip())
        else:
            name.append("N/A")  # If name is missing, append "N/A"
        
        # Scrape product price
        product_price = i.find('span', class_='a-price-whole')
        if product_price:
            price.append(product_price.text.strip())
        else:
            price.append("N/A")  # If price is missing, append "N/A"

# Create a DataFrame from the scraped data
dict1 = {'name': name, 'price': price}
df1 = pd.DataFrame(dict1)

# Display the DataFrame
df1
```

```python
df1.to_csv('amazon_whey_protein.csv', index=False)
```
